```python
import pandas as pd

data = pd.read_csv("students.csv")
print(data)

average = data['Marks'].mean()
print("Average Marks:", average)

passing = data[data['Marks'] >= 75]
print("Number of Passing Students:", len(passing))

top = data['Marks'].max()
top_student = data[data['Marks'] == top]['Name'].values[0]
print("Top Scorer:", top_student)
```

```python
import matplotlib.pyplot as plt

plt.bar(data['Name'], data['Marks'])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance Analysis")
plt.show()
```

```python
pass_count = len(data[data['Marks'] >= 75])
fail_count = len(data[data['Marks'] < 75])

plt.pie([pass_count, fail_count], labels=['Pass', 'Fail'], autopct='%1.1f%%')
plt.title("Pass vs Fail Students")
plt.show()
```
